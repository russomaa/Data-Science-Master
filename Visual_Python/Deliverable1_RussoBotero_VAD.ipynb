### ¿Cómo se distribuyen las causas de crímenes en España ? ¿Se distribuyen igual en los paises vecinos ?

```python
#We preapare dataset with only Spain info
crimes_esp = crimes[crimes.country.isin(['ES'])] #pick out only Spain data
crimes_esp = crimes_esp.drop([208]) #erase item with totals
crimes_esp.head() #we check our new table
```

```python
#Data with types of crimes and total count 
crimes_esp.drop(crimes.columns[[0, 2]], axis=1, inplace=True) #we get rid of columns we don't need 
total=crimes_esp.iloc[:,1:].sum(axis = 1) #add up yearly info 
crimes_esp['TOTAL'] = total  #we get our total column
crimes_esp_total = crimes_esp[['iccs', 'TOTAL']] 
```

```python
crime_type = crimes_esp_total["iccs"]
crime_total = crimes_esp_total["TOTAL"]
labels = crime_type.values
sizes = crime_total.values
colors = ["#8FBC8F", "#3CB371", "#808000", "#5F9EA0", "#008B8B","#3366FF"]
explode = (0.01,0.01,0.01,0.01,0.01,0.01)
fig1,ax1 = plt.subplots(figsize=(10, 7))
fig1.subplots_adjust(0.3,0,1,1)
ax1.pie(sizes, explode=explode, colors = colors, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=100)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Distribución de crimines por causa en España, 1993-2012" ,fontsize=14, y= 1.05)
plt.axis('equal')
plt.show()
```

La base de datos empleada recoge el registro de crímenes referente a 6 categorías, homicidio, crimen violento, robo doméstico, robo, robo de un vehículo motorizado, tráfico de drogas.
Con respecto a la distribución del crimen en España, y gracias a la gráfica anterior, se puede apreciar como en el periodo observado (1993-2012), el crimen registrado ha seguido la siguiente distribución:

1º Robo de vehículos motorizados

2º Crimen violento

3º Robo

4º Robo doméstico

5º Tráfico de droga

6º Homicidio

### ¿Cómo ha evolucionado el total de crímenes en España y en los paises vecinos?

```python
crimes = pd.read_csv('crim_gen.tsv', sep=',', na_values=': ')
crimes_esp_pt_fr = crimes[crimes.country.isin(['ES', 'PT','FR'])]
crimes_esp_pt_fr.head()
#crimes_esp_pt_fr.columns
```

```python
total_crimes_neigh = crimes_esp_pt_fr.query('iccs == "TOTAL"').values
total_crimes_neigh 
```

```python
x = np.arange(len(total_crimes_neigh[0][3:]))
plt.figure(figsize=(20,10))
plt.plot(x, total_crimes_neigh[0][3:]/1e6,label='España',color="#2E8B57")
plt.plot(x, total_crimes_neigh[1][3:]/1e6,label='Francia',color="#5F9EA0")
plt.plot(x, total_crimes_neigh[2][3:]/1e6,label='Portugal',color="#DB7093")



plt.xticks(x, crimes_esp_pt_fr.columns[3:], rotation=40)
plt.xlabel('Año',fontsize=15)
plt.ylabel('Número de crímenes reportados(millones)',fontsize=15)
plt.tight_layout()
plt.title("Evolución del número de crímenes, 2012-1993" ,fontsize=20, y= 1.05)
plb.legend(loc='upper left',fontsize=15)
plt.show()
```

Con respecto a la evolución del crimen en los años observados, si bien es difícil apreciar la situación particular de cada país, en particular el caso de Portugal por su tamaño en comparación con España y Francia. De todamas maneras, se puede ver como el registro del crimen se ha mantenido bastante estable (0.5 millones de crímenes reportados anualmente).

El caso de España, se puede observar una ligera tendencia creciente desde 1993-2000, con un valle en 2000 en los 2.0 millones, y otra vuelta al alza que se estabiliza en 2002, manteniendo el número de crímenes reportados al rededor de 2.5 millones cada año.

Francia, muestra cierta volatilidad más con periodos pronunciados de altas y bajas en el número de registros anuales. Los últimos registros permiten apreciar unos 3.5 millones de crímenes reportados, en comparación con el pico más alto que llegó a rondar los 4.0 millones durante los primeros años de la década 2000. 

### ¿Cúales son los 4 paises más peligrosos?  

En base a la información de la cual se dispone, se determina que los países más "peligrosos" serán los que tengan un registro de crimen global más elevado en comparación con el resto de países, sin tomar en cuenta tamaño, ni población. 

```python
#We preapare data to determine 4 most dangerous countries based on absolute values 
total_crimes = crimes[crimes['iccs'] == 'TOTAL'] #we only keep the 'total' rows for each country
total_crimes['total'] = total_crimes.loc[:,'2012':'1993'].sum(skipna=True, axis=1) #we add up all our yearly data
total_crimes = total_crimes.loc[:, ['country','total']]
total_crimes = total_crimes.sort_values(by=['total'],ascending=False)[:5] 
total_crimes = total_crimes.drop([211]) #erase FX item, metropolitan France, appears to show the same data 
total_crimes
```

```python
import matplotlib.cm as cm #to get set of colormaps 
from matplotlib.colors import Normalize #data scaling
#https://matplotlib.org/2.0.2/api/cm_api.html#module-matplotlib.cm

index = np.arange(len(total_crimes.country))
values = total_crimes.total.values

#build color map 
my_cmap = cm.get_cmap('jet')
my_norm = Normalize(vmin=0, vmax=4)

plt.figure(figsize=(20,10))
plt.bar(index, values, color=my_cmap(my_norm(index)),  edgecolor='#F0E68C')
plt.xlabel('Paises', fontsize=15)
plt.ylabel('Número de crímenes', fontsize=15)
plt.xticks(index, total_crimes.country.values, fontsize=15, rotation=15)
plt.title('Los 4 países con el registro de crímenes más altos, 1993-2012',fontsize=20, y= 1.02)
plt.show()
```

Como bien se puede observar, los 4 países con mayor registro de crimen en los años observados (1993-2012) son: Estados Unidos, Alemania, Inglaterra&Gales, y Francia.
En este caso se han tomado los valores totales absolutos (la media se ha calculado también, obteniendo los mismos resultados, al igual que teniendo en cuenta Escocia e Irlanda del Norte, UK se mantiene tercero). 

## **Nutrients:**

Importación de datos

```python
pd.options.display.max_rows = 100
db = json.load(open('foods-2011-10-03.json'))
```

```python
# Exploramos los nutrientes 
nutrients = pd.DataFrame(db[0]['nutrients'])
print(nutrients[:10])

# Data Frame con la info importante asociada a los nutrientes (se filtra el resto: tags, portions,...)
info_keys = ['description', 'group', 'id', 'manufacturer']
info = pd.DataFrame(db, columns=info_keys)

print("Info: ", info[:10])
pd.options.display.max_rows = 100
pd.value_counts(info.group)

# Lista de Data Frame de nutrientes
nutrients = []
for rec in db:
    fnuts = pd.DataFrame(rec['nutrients'])
    fnuts['id'] = rec['id']
    nutrients.append(fnuts)

nutrients = pd.concat(nutrients, ignore_index=True)
print(nutrients[:10])
```

```python
nutrients.duplicated().sum()
nutrients = nutrients.drop_duplicates()

col_mapping = {'description' : 'food', 'group' : 'fgroup'}
info = info.rename(columns=col_mapping, copy=False)
col_mapping = {'description' : 'nutrient', 'group' : 'nutgroup'}
nutrients = nutrients.rename(columns=col_mapping, copy=False)
ndata = pd.merge(nutrients, info, on='id', how='outer') #i'm going to need this for food groups 
```

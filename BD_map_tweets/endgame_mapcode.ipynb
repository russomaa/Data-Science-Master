El objetivo de este trabajo es capturar y procesar Tweets. 
Durante una semana se han capturado tweets con el hashtag #AvengersEndGame.
Esta hashtag hace referencia a la más reciente entrega de las películas de Marvel Superheroes, la  cual dará cierre a la primera fase del MCU trás su comienzo hace 11 años.
En el mapa se pintaran los tweets en base a su origen (longitud y latitud), con el fin de observar donde  existe una mayor conglomeración de estos, durante la semana de estreno.<br>
**Nota**: Dado que la geo localización de los tweets no es información compulsoria, la cantidad de tweets que se pueden obtener con estos parámetros es reducida. 

```python
import pandas as pd 
import numpy as np
import folium
import geocoder
#https://georgetsilva.github.io/posts/mapping-points-with-folium/
```

```python
#data
data = pd.read_csv("data/coordenadas.csv") 
data.head()
```

```python
locations=data[['latitude','longitude']]
locationslist=locations.values.tolist()
len(locationslist)
locationslist[7]
```

```python
endgame_map = folium.Map(location=[51.5079, 0.0877],tiles='Stamen Toner',zoom_start=2)
for point in range(0, len(locationslist)):
    folium.Marker(locationslist[point], popup=data['user'][point],icon=folium.Icon(icon='star',color='green')).add_to(endgame_map)        
```
